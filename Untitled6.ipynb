```python
import timeit
import random
import matplotlib.pyplot as plt
import math

def binary_search(arr, target, low, high):
    if high >= low:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, target, low, mid - 1)
        else:
            return binary_search(arr, target, mid + 1, high)
    return -1  

def measure_time(n_values, iterations=50):
    times = []
    for n in n_values:
        arr = list(range(n))  
        target = arr[-1]  
        time_taken = timeit.timeit(lambda: binary_search(arr, target, 0, len(arr) - 1), number=iterations)
        avg_time = time_taken / iterations  
        times.append(avg_time)

    return times

n_values = [1000, 2000, 5000, 10000, 20000, 50000, 100000]
times = measure_time(n_values)


scaling_factor = times[-1] / math.log2(n_values[-1])
theoretical_times = [scaling_factor * math.log2(n) for n in n_values]

plt.figure(figsize=(8, 5))
plt.plot(n_values, theoretical_times, marker='s', color='r', linestyle='-', label="Theoretical O(log n)")

plt.title("Binary Search: Time Complexity Graph")
plt.xlabel("Number of Elements (n)")
plt.ylabel("Time Taken (seconds)")
plt.grid(True)

plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
def measure_time(sorting_function, n_values):
    times = []
    for n in n_values:
        arr = [random.randint(1, 10000) for _ in range(n)]
        start_time = time.time()
        sorting_function(arr)
        end_time = time.time()
        times.append(end_time - start_time)
    return times


n_values = [100, 500, 1000, 5000, 10000, 20000, 50000]
insertion_sort_times = measure_time(insertion_sort, n_values[:4])  
heap_sort_times = measure_time(heap_sort, n_values)
plt.plot(n_values[:4], insertion_sort_times, marker='o', label="Insertion Sort", color='r')
plt.plot(n_values, heap_sort_times, marker='o', label="Heap Sort", color='b')
plt.title("Sorting Time vs List Size")
plt.xlabel("Number of Elements (n)")
plt.ylabel("Time Taken (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from collections import deque
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    while queue:
        node = queue.popleft()
        print(node, end=" ")
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("BFS Traversal starting from node 'A':")
bfs(graph, 'A')
```

```python
def dfs(graph,node, visited=None):
    if visited is None: 
        visited = set()
    visited.add(node)
    print(node,end=" ")
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph,neighbor,visited)
graph = {
        'A':['B','C'],
        'B':['A','D','E'],
        'C':['A','F'],
        'D':['B'],
        'E':['B','F'],
        'F':['C','E']
}
print("DFS Traversal starting from Node 'A':")
dfs(graph,'A')
```

```python
import heapq
def dijkstra(graph, start):
    distances = {vertex: float('inf') for vertex in graph}
    distances[start] = 0
    priority_queue = [(0, start)]
    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)
        if current_distance > distances[current_vertex]:
            continue
        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
        return distances

graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

start_vertex = 'A'
shortest_paths = dijkstra(graph, start_vertex)

print(f"Shortest paths from vertex '{start_vertex}':")
for vertex, distance in shortest_paths.items():
    print(f"Distance from {start_vertex} to {vertex}: {distance}")
```

```python
import heapq

def prim_mst(graph):
    mst = []
    visited = set()
    min_heap = [(0, list(graph.keys())[0])]
    
    while min_heap:
        cost, vertex = heapq.heappop(min_heap)
        
        if vertex in visited:
            continue
        
        visited.add(vertex)
        
        if cost != 0:
            mst.append((prev_vertex, vertex, cost))
        
        for neighbor, weight in graph[vertex].items():
            if neighbor not in visited:
                heapq.heappush(min_heap, (weight, neighbor))
        
            prev_vertex = vertex
    
    return mst

graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3}
}

mst = prim_mst(graph)

print("Edges in the Minimum Spanning Tree (MST):")
total_cost = 0
for edge in mst:
    print(f"{edge[0]} - {edge[1]} with cost {edge[2]}")
    total_cost += edge[2]
print(f"\nTotal cost of the Minimum Spanning Tree: {total_cost}")
```

```python
def floyd_warshall(graph):
    vertices = list(graph.keys())
    dist = {vertex: {v: float('inf') for v in vertices} for vertex in vertices}
     
    for vertex in vertices:
        dist[vertex][vertex] = 0 
    for vertex in graph:
        for neighbor, weight in graph[vertex].items():
            dist[vertex][neighbor] = weight
    for k in vertices:
        for i in vertices:
            for j in vertices:
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
     
    return dist
graph = {
     'A': {'B': 4, 'C': 2},
     'B': {'A': 4, 'C': 5, 'D': 10},
     'C': {'A': 2, 'B': 5, 'D': 3},
     'D': {'B': 10, 'C': 3}
}
shortest_paths = floyd_warshall(graph)
print("Shortest paths between all pairs of vertices:")
for vertex in shortest_paths:
    print(f"{vertex}: {shortest_paths[vertex]}")
```

```python
def warshall(graph):
    vertices = list(graph.keys())
    n = len(vertices)
    reach = [[0] * n for _ in range(n)]
    vertex_to_index = {vertex: idx for idx, vertex in enumerate(vertices)}
    
    for i in range(n):
        reach[i][i] = 1
        for neighbor in graph[vertices[i]]:
            reach[i][vertex_to_index[neighbor]] = 1
    
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if reach[i][k] == 1 and reach[k][j] == 1:
                    reach[i][j] = 1
    
    return reach

def print_transitive_closure(reach):
    print("Transitive Closure Matrix:")
    for row in reach:
        print(row)

graph = {
    'A': ['B', 'C'],
    'B': ['C'],
    'C': ['D'],
    'D': []
}

transitive_closure = warshall(graph)
print_transitive_closure(transitive_closure)
```

```python
def find_min_max(arr, low, high):
    if low == high:
        return arr[low], arr[low]
    elif high == low + 1:
        if arr[low] < arr[high]:
            return arr[low], arr[high]
        else:
            return arr[high], arr[low]
    
    mid = (low + high) // 2
    left_min, left_max = find_min_max(arr, low, mid)
    right_min, right_max = find_min_max(arr, mid + 1, high)
    
    return min(left_min, right_min), max(left_max, right_max)

arr = [12, 3, 5, 7, 19, 1, 10]
n = len(arr)
minimum, maximum = find_min_max(arr, 0, n - 1)
print("Minimum:", minimum)
print("Maximum:", maximum)
```
